```python
dtype_names = [
]

import pandas.api.types as ptypes

ALLOWED_TYPES = [
    'bool',
    'category',
    'complex64',
    'complex128',
    'datetime64',
    'datetime64[ns]',
    'datetime64[ns, UTC]',
    'float16',
    'float32',
    'float64',
    'int8',
    'int16',
    'int32',
    'int64',
    'object',
    'string',
    'timedelta',
    'timedelta64',
    'uint8',
    'uint16',
    'uint32',
    'uint64',

]

def switchTarget(dataframe, column, dtype):
  try:

    if dtype in ALLOWED_TYPES:
      dataframe[column] = dataframe[column].astype(dtype)
    else:
      raise Exception(f"{dtype} is not an acceptable pandas type")
  except Exception as error:
    raise error
  
```

## Spliting the values in the charstat column to individual columns


#### New columns
- correct
- incorrect
- extra (I think these are the extra characters on each word?!)
- missed (the missed key/charcter on each word). -> great to know my weak fingers and to work on them

```python
df.charStats.head(5)
```

```python
def cleanCharSets():
    try:
        # Create new columns if they do not already exist
        for col in ['correct', 'incorrect', 'extra', 'missed']:
            if col not in df.columns:
                df[col] = None
        
        # Iterate over each row
        for index, row in df.iterrows():
            target = row['charStats']
            splits = target.split(";")
            
            # Ensure splits have the expected length
            if len(splits) == 4:
                correct, incorrect, extra, missed = splits

                # Populate the DataFrame with each value
                df.at[index, 'correct'] = correct
                df.at[index, 'incorrect'] = incorrect
                df.at[index, 'extra'] = extra
                df.at[index, 'missed'] = missed

        return df
    except Exception as error:
        raise error


```

```python
new_df = cleanCharSets()
new_df.head(5)
```

```python
# new_df.drop(columns="charStats", inplace=True)
new_df.dropna()
new_df.head(5)
typing_metrixs_corr = new_df[["correct", "incorrect", "extra", "missed", "consistency", "acc", "wpm", "rawWpm", "mode2", "restartCount"]].corr()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Generate a large random dataset
mask = np.triu(np.ones_like(typing_metrixs_corr, dtype=bool))

f, ax = plt.subplots(figsize=(6, 3))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(typing_metrixs_corr  , mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})



```
